```python
def convertfloat(x):
    try:
        return float(x)
    except:
        return -999.99
fltdata = csvdata.map(lambda x: x.split(",")).map(lambda x: [a for a in x[:5]]+[convertfloat(b) for b in x[5:]])
fltdata
```

```python
r = sc.parallelize(((1,2), (3,4), (1,4), (5,3), (6,2), (4,5), (1,7)))
seop = (lambda x, y : (x[0]+y, x[1]+1, x[2]+1/2))
comop = (lambda x, y: (x[0]+y[0], x[1]+y[1], x[2]-y[2]))
x = r.map(lambda x: x[-1]).aggregate((0,0,0), seop, comop)
x
```

```python
seop1 = (lambda x, y : (x[0]+y, x[1]+1))
comop1 = (lambda x, y: (x[0]+y[0], x[1]+y[1]))
ru = r.union(r)
y = ru.map(lambda x: (x[0], x[-1])).aggregateByKey((0,0), seop1, comop1)
(r.collect(), y.collect())
```

```python
ru.getNumPartitions()
```

```python
ru_part3 = ru.coalesce(3)
ru_part3.getNumPartitions()
```

```python
ru_part3.glom().collect()
```
